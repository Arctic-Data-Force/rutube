```python
import pandas as pd
import numpy as np


from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
v = pd.read_parquet('/kaggle/input/axaxaxaxa/videos.parquet')
u = pd.read_csv('/kaggle/input/axaxaxaxadb/merged_df.csv')
```

```python
v.info()
u.info()
```

```python
v = v[['item_id', 'category_title']]
u = u[['user_id', 'item_id', 'is_watched', 'is_autorized']]
```

```python
df = u.merge(v, on='item_id', how='left')
```

```python
df
```

```python
df['category_title'].unique()
```

```python
merged_df = pd.get_dummies(df, columns=['category_title'], prefix='category')
```

```python
grouped_df = merged_df.groupby('user_id').sum().reset_index()
```

```python
grouped_df
```

```python
grouped_df2 = grouped_df.drop(['item_id', 'is_watched', 'is_autorized'], axis=1)
grouped_df3 = grouped_df2.copy()
```

```python
grouped_df2['total_watch_time'] = grouped_df2.iloc[:, 1:].sum(axis=1)
```

```python
grouped_df2
```

```python
for category_column in grouped_df2.columns[1:-1]:
    preference_column = 'preference_' + category_column
    grouped_df2[preference_column] = grouped_df2[category_column] / grouped_df2['total_watch_time']
```

```python
grouped_df2
```

```python
preference_df = grouped_df2[['user_id'] + [col for col in grouped_df2.columns if col.startswith('preference_')]]
```

```python
preference_df.columns = ['user_id'] + [col.replace('preference_', '') for col in preference_df.columns[1:]]
```
